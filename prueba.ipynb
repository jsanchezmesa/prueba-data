```python
import requests

url = 'https://es.wikipedia.org/wiki/Copa_Mundial_de_F%C3%BAtbol_de_2018'
html = requests.get(url).content
```

```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(html, "lxml")

#Get teams
teams = [e.text.strip() for e in soup.select('.vevent tbody td[width="22%"]')]
#Filter local teams
localTeams = teams[0::2]
#Get local goals
goals = [int(e.text.strip()[0]) for e in soup.select('.vevent tbody td[width="12%"]')]

localGoals = {}

# match local teams with goals
for index in range(0, len(localTeams)):
    if localTeams[index] in localGoals:
        localGoals[localTeams[index]] += goals[index]
    else:
        localGoals[localTeams[index]] = goals[index]

# sort by goals
sortedGoals = sorted(localGoals.items(), key=lambda elem: elem[1], reverse=True)

# print table
pd.DataFrame(sortedGoals, columns=["Team","Goals"])
```

```python
import re
players = [ e.text.strip()
           .replace("(pen.),", ",")
           .replace("(pen.)", "\n")
           .replace("(a.g.)", "(a.g.)\n") for e in soup.select('td[align="right"]')]
#get only goals, not teams
players = filter(lambda elem: '\'' in elem,players)

#split player from other players
playersSplit = []
for elem in players:
    a = re.split("\'[A-Z]", elem)
    for player in a:
        if(player.count('(a.g.)') == 0):
            playersSplit.append(player)


playerGoal = {}
for i in playersSplit:
    tmp = i.split(" ")[0]
    if tmp in playerGoal:
        playerGoal[tmp] += i.count(',')+1
    else:
        playerGoal[tmp] = i.count(',')+1
        
sortedPlayerGoals = sorted(playerGoal.items(), key=lambda elem: elem[1], reverse=True)
pd.DataFrame(sortedPlayerGoals, columns=["Player","Goals"])
```

```python
#Get final rounds goals
finalRoundGoals = goals[-16::2]
avgFinalGoals = sum(finalRoundGoals)/len(finalRoundGoals)
    
print(str(avgFinalGoals) + ' average goals per match above quarter finals')
```
